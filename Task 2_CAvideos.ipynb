```python
#First, import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#import the file and create a data frame
vid = pd.read_csv('CAvideos.csv')

vid.head()
```

```python
#HISTOGRAM
#It is good to specify the columns. It will produce histograms of each of the specified columns.

df = pd.DataFrame(vid, columns = ['category_id', 'views', 'likes', 'dislikes', 'comment_count'])
df.hist()

plt.style.use('ggplot') # make graphs in the style of R's ggplot

#show plots
plt.show()
```

```python
#Interpretation
# The y-axis represents the count of the feature. The x-axis is a range, from initial point of the bar to the end of the bar
# In category id, the category of the id fall mainly between 22 and 25. None is more than 30 and there are some in lower ranges.
```

```python
# to show selected box plots on the same chart
data_to_plot = [df['likes'], df['dislikes'], df['comment_count'], df['category_id']]

#Create a fig instance
fig = plt.figure(1, figsize=(9, 6))

#Create axes instance
ax = fig.add_subplot(111)

#Create the boxplot
bp = ax.boxplot(data_to_plot)

#Custom x-axis labels
ax.set_xticklabels(['likes', 'dislikes', 'comment_count', 'category_id'])

# If you want to save it in the directory
#fig.savefig('fig1.png', bbox_inches='tight')
```

```python
#SCATTER PLOTS
plt.scatter(df['comment_count'], df['dislikes'])
plt.xlabel('no of comment')
plt.ylabel('no of dislikes')
plt.show()

#scatter plot between dislike and category id
plt.scatter(df['category_id'], df['dislikes'])
plt.xlabel('category')
plt.ylabel('no of dislikes')
plt.show()

#scatter plot between dislike and views
plt.scatter(df['views'], df['dislikes'])
plt.xlabel('no of views')
plt.ylabel('no of dislikes')
plt.show()

#scatter plot between like and views
plt.scatter(df['views'], df['likes'])
plt.xlabel('no of views')
plt.ylabel('no of likes')
plt.show()

#scatter plot between like and category id
plt.scatter(df['category_id'], df['likes'])
plt.xlabel('category')
plt.ylabel('no of likes')
plt.show()

#scatter plot between like and comment
plt.scatter(df['likes'], df['comment_count'])
plt.xlabel('no of likes')
plt.ylabel('no of comments')
plt.show()
```
